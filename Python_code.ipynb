```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
df = pd.read_csv('visits_sample.csv')
```

```python
df.rename({' visits': 'visits'},axis=1, inplace=True)
```

```python
def rem_space(string):
    return string[1:]
```

```python
df.visits = list(map(rem_space,df.visits))
```

```python
df
```

```python
date = datetime.date(2018, 1, 1)  ##choose this date because 2018 starts with Monday
```

```python
days_per_day = []
for i in df.visits:
    counts = {}
    i = i.split(" ")
    for j in i:
        j = int(j)
        curr_date = date + datetime.timedelta(days=j-1)
        curr_day = curr_date.strftime("%A")
        if curr_day in counts:
            counts[curr_day] += 1
        else:
            counts[curr_day] = 1
    days_per_day.append(counts)
```

```python
dpd_df = pd.DataFrame.from_dict(days_per_day)
```

```python
dpd_df['visitor_id'] = df['visitor_id']
```

```python
dpd_df = dpd_df[['visitor_id', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]
```

```python
dpd_df
```

```python
days_per_month = []
for i in df.visits:
    counts = {}
    i = i.split(" ")
    for j in i:
        j = int(j)
        curr_date = date + datetime.timedelta(days=j-1)
        curr_day = curr_date.strftime("%B")
        if curr_day in counts:
            counts[curr_day] += 1
        else:
            counts[curr_day] = 1
    days_per_month.append(counts)
```

```python
dpm_df = pd.DataFrame.from_dict(days_per_month)
```

```python
dpm_df = dpm_df[['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
       'September', 'October', 'November', 'December']]
```

```python
dpm_df
```

```python
dfx = pd.concat([dpd_df,dpm_df],axis=1)
```

```python
days_per_year = []
for i in df.visits:
    counts = {}
    i = i.split(" ")
    for j in i:
        j = int(j)
        curr_date = date + datetime.timedelta(days=j-1)
        curr_day = curr_date.strftime("%Y")
        if curr_day in counts:
            counts[curr_day] += 1
        else:
            counts[curr_day] = 1
    days_per_year.append(counts)
```

```python
dpy_df = pd.DataFrame.from_dict(days_per_year)
```

```python
dpy_df
```

```python
dfx = pd.concat([dfx,dpy_df],axis=1)
```

```python
days_per_weekno = []
for i in df.visits:
    counts = {}
    i = i.split(" ")
    for j in i:
        j = int(j)
        curr_date = date + datetime.timedelta(days=j-1)
        curr_day = curr_date.strftime("%W")
        if curr_day in counts:
            counts[curr_day] += 1
        else:
            counts[curr_day] = 1
    days_per_weekno.append(counts)
```

```python
dpw_df = pd.DataFrame.from_dict(days_per_weekno)
```

```python
dpw_df = dpw_df[dpw_df.columns.sort_values().to_list()]
```

```python
dpw_df = dpw_df.add_prefix('week-')
```

```python
dpw_df
```

```python
dfx = pd.concat([dfx,dpw_df],axis=1)
```

```python
average_days = []
for i in df.visits:
    counts = []
    i = i.split(" ")
    i = list(map(int, i))
    for j in range(len(i)):
        try:
            counts.append(i[j+1] - i[j])
        except:
            total = sum(counts)/len(counts)
    average_days.append(round(total))
```

```python
dfx['Average_app_open_days'] = average_days
```

```python
dfx['Average_app_open_days']
```
